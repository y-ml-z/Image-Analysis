# **Image Analysis**




```
import cv2
import os
from matplotlib import pyplot as plt
%matplotlib inline
```

```
img = cv2.imread("tum_m.jpg")
```

```
plt.imshow(img)
```

```
# Image Processing using OpenCV

img_corrected = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.axis("off") #remove axes ticks
plt.imshow(img_corrected)
```

```
#Change the size of the displayed image
from pylab import rcParams

rcParams['figure.figsize'] = 15, 14

plt.axis("off") #remove axes ticks
plt.imshow(img_corrected)
```

```
#Convert to Grayscale
new_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(new_img, cmap = 'gray')
plt.axis("off") #remove axes ticks
plt.title('Grayscale Image')
```

```
#Canny Edge Detection
rcParams['figure.figsize'] = 10, 12


edges = cv2.Canny(img_corrected, 
                  threshold1=1000, 
                  threshold2=50)

plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])

```

```
#Analyzing Images using Histograms
#Histograms - Grayscal

  
rcParams['figure.figsize'] = 10,7

plt.hist(new_img.ravel(),256,[0,256],color="green")
plt.title('Histogram of Grayscale TUM_M.jpg')
plt.show()
```

```
 #RGB histogram
 rcParams['figure.figsize'] = 8, 4

color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```

###**Interpretation:**
In the histogram above, you can observe that:
- of the lighter pixels (around `x >= 200`), there are more red pixels than green or blue pixels. This is likely because of the redness in the faces of the persons.
- there seems to be a similar amount of dark red, dark blue, and dark green pixels (around `x<50`).
- in terms of middle-tones, there seems to be more blue pixels than green or red.
